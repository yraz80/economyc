```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
% matplotlib inline
plt.style.use('ggplot')
```

```python
dframe = pd.read_csv('.././data/personal2016_2.csv')
dframe['total'] = dframe['men'] + dframe['woman']
dframe['total_sm'] = dframe[['small_max', 'micro_max', 'medium_max']].sum(axis=1).astype(int)
dframe
```

```python
df = dframe.iloc[1:, 0:].copy()
df = df.sort_values('total',ascending=False)
df
```

```python
other = df.total.sum() - df.total_sm.sum()
print('\nУкрупненные показатели по занятости в 2016 году:\n')
print('Общая занятость                                     :', round(df.total.sum()/1000, 2), ' млн.чел.')
print('Занятость на предприятиях малого и среднего бизнеса :', round(df.total_sm.sum()/1000, 2), ' млн.чел.')
print('Занятость на крупных предприятиях и самозанятых     :', round(other/1000, 2),
      ' млн.чел')
print('Итого самозанятых                                   :', 25.00, ' млн.чел')
print('Итого на крупных предприятиях                       :', round(other/1000 - 25, 2), ' млн.чел')
```

```python
plt.figure(figsize=(14, 8))

plt.title('\nРаспределение занятых по отраслям\n')
plt.bar(df.name, df.total, label='Занятость всего в экономике России {} млн. чел.'.format(df.total.sum()/1000))
plt.bar(df.name, df.total_sm, label='Занятость в малом и среднем бизнесе {} млн. чел.'.format(df.total_sm.sum()/1000))

dct = dict(df[['name', 'total']].values)

for i, l in dct.items():
    plt.text(i, l+400, l, ha='center')

dct = dict(df[['name', 'total_sm']][:-1].values)

for i, l in dct.items():
    plt.text(i, l+400, l, ha='center', color='w')

plt.ylim(-1000, df['total'].max() + df['total'].max()*0.2)

plt.legend()
plt.xticks(rotation=90)
plt.show()
```

```python
df['total_all'] = df['total'] - df['total_sm']

plt.figure(figsize=(14, 8))

plt.title('\nРаспределение занятых по отраслям\n')
plt.bar(df.name, df.total_all, label='Занятость без учета малого и среднего бизнеса {} млн. чел.'.format(
    df.total_all.sum()/1000))

dct = dict(df[['name', 'total_all']].values)

for i, l in dct.items():
    plt.text(i, l+400, l, ha='center')


plt.ylim(-1000, df['total_all'].max() + df['total_all'].max()*0.2)

plt.legend()
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(14, 8))

plt.title('\nЗанятость по отраслям\n')
plt.plot(df.name, df.woman, label='Женщины')
plt.plot(df.name, df.men, label='Мужчины')

plt.scatter(df.name, df.woman)
plt.scatter(df.name, df.men)
plt.xticks(rotation=90)

dct = dict(df[['name', 'woman']].values)

for i, l in dct.items():
    plt.text(i, l+200, l, ha='right')
    

dct = dict(df[['name', 'men']].values)

for i, l in dct.items():
    plt.text(i, l-300, l, ha='right')
    
plt.ylim(0, df['woman'].max() + df['woman'].max()*0.1)
plt.legend()
plt.show()
```
